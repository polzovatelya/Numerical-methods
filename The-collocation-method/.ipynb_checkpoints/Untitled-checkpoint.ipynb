```python
def y0_p(u0, u1, x):
    return u0 + (u1 - u0)*x
def yn_p(x, n):
    return x**n*(1 - x)
def dyn_p(x, n):
     return n*x**(n-1) - (n+1)*x**n
def ddyn_p(x, n):
     return n*(n-1)*x**(n-2) - (n+1)*n*x**(n-1)


def y0_t(u0, u1, x):
    return u1*np.sin(np.pi/2*x) + u0*np.cos(np.pi/2*x)
def yn_t(x, n):
     return n* np.sin(np.pi*n*x)
def dyn_t(x, n):
     return n* np.pi*n*np.cos(np.pi*n*x)
def ddyn_t(x, n):
     return -n* (np.pi*n)**2*np.sin(np.pi*n*x)
```

```python
def L0_p(x, n, k, q, u0, u1):
    return q*y0_p(u0, u1, x)
def Ln_p(x, n, k, q):
     return -k*ddyn_p(x, n) + q* yn_p(x, n)
    
def L0_t(x, n, k, q, u0, u1):
     return -k*(-u1*(np.pi/2)**2*np.sin(np.pi/2*x) - u0*(np.pi/2)**2*np.cos(np.pi/2*x)) + q
def Ln_t(x, n, k, q):
     return -k*ddyn_t(x, n) + q*yn_t(x, n)
```

```python

import numpy as np
from scipy.linalg import solve

def f(x):
    return 3*x**4 - 15*x**3 + 15*x**2 - 39*x + 8
n = 5
k = -1
q = 3
u0 = 2
u1 = -4

collocation_points = np.linspace(0, 1, n)[1:-1]
print(collocation_points)

#       
A_poly = np.zeros((n-2, n-2))
A_trig = np.zeros((n-2, n-2))
b = np.zeros(n-2)

#       
for i, x in enumerate(collocation_points):
    for j in range(1, n-1):
        A_poly[i, j-1] = Ln_p(x, j, k, q)
    b[i] = f(x) - L0_p(x, n, k, q, u0, u1)
    
#       
coefficients_poly, _, _, _ =  np.linalg.lstsq(A_poly, b, rcond=None)
print(coefficients_poly)

#      
def y_poly(x):
    y = y0_p(u0, u1, x)
    for i, c in enumerate(coefficients_poly):
        y += c * yn_p(x, i+1)
    return y

#       
for i, x in enumerate(collocation_points):
    for j in range(1, n-1):
        A_trig[i, j-1] = Ln_t(x, j, k, q)
    b[i] = f(x) - L0_t(x, n, k, q, u0, u1)
#       
coefficients_trig, _, _, _ =  np.linalg.lstsq(A_trig, b, rcond=None)
print(coefficients_trig)
#      
def y_trig(x):
    y = y0_t(u0, u1, x)
    for i, c in enumerate(coefficients_trig):
        y += c * yn_t(x, i+1)
    return y

#  
print(" :")
for x in np.linspace(0, 1, 10):
    print(f"x={x:.2f}, y(x)={y_poly(x):.4f}")

print("\n :")
for x in np.linspace(0, 1, 10):
    print(f"x={x:.2f}, y(x)={y_trig(x):.4f}")
```
